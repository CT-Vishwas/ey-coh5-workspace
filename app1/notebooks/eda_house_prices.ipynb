Exploratory Data Analysis (EDA) is a crucial step in the data analysis process that involves summarizing the main characteristics of a dataset, often using visual methods. The goal of EDA is to gain insights into the data, identify patterns, detect anomalies, and formulate hypotheses for further analysis.Here are some common techniques and tools used in EDA:
1. **Summary Statistics**: Calculating measures such as mean, median, mode, standard deviation, and percentiles to understand the central tendency and dispersion of the data.
2. **Data Visualization**: Creating various types of plots and charts to visualize the distribution of the data, relationships between variables, and trends over time. Common visualizations include histograms, box plots, scatter plots, and bar charts.
3. **Correlation Analysis**: Assessing the relationships between variables using correlation coefficients and heatmaps to identify potential associations.

Average Price by the postal code. Which postal code has the highest average price?

```python
# Group by postal code and calculate the average price for each postal code
avg_price_by_postal_code = df.groupby('Postal Code')['Price'].mean()
avg_price_by_postal_code.sort_values(ascending=False)
```

```python
# Plotting the average price by postal code
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_price_by_postal_code.index, y=avg_price_by_postal_code.values)
plt.xticks(rotation=90)
plt.xlabel('Postal Code')
plt.ylabel('Average Price')
plt.title('Average Price by Postal Code')
plt.show()
```
