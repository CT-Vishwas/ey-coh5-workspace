Fix Missing Data, Formats

```python
# Fix Missing Data, Formats
data = {
    "A": [1, 2, None, 4],
    "B": ["2021-01-01", "2021-02-01", "2021-03-01", None],
    "c": ["10", "20", "30", "40"]
}
sample_df = pd.DataFrame(data)
sample_df
```

```python
sample_df.info()
```

Eiter yoither fill the missing data or drop the rows/columns with missing data. Also, convert the data types to appropriate formats.

```python
sample_df1 = sample_df.copy()
sample_df1.dropna(inplace=True)
sample_df1
```

```python
sample_df2 = sample_df.copy()
sample_df2.fillna({
    "A": sample_df2["A"].mean(),
    "B": pd.to_datetime(sample_df2["B"].fillna("2021-01-01")),
    "c": sample_df2["c"].fillna("0")
}, inplace=True)
sample_df2
```

```python
sample_df2['c'] = sample_df2['c'].astype(int)
sample_df2.info()
```
